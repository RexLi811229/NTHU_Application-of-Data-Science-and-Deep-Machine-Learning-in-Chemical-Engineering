```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
from utils import get_player_data, plot_decision_regions
```

```python
"""Plot sigmoid function"""
x = np.linspace(-5, 5, 1000)
y = 1 / (1 + np.exp(-x))
plt.plot(x, y)
plt.xlim(-5, 5)
plt.ylim(0, 1)
plt.savefig('ML_images/sigmoid')
plt.show()
```

```python
"""Player position prediction by logistic regression"""
X_player_n, y_player = get_player_data()

lr = LogisticRegression(C=1.0, random_state=1, solver='lbfgs', multi_class='multinomial')
lr.fit(X_player_n, y_player)
plot_decision_regions(X_player_n, y_player, classifier=lr)
plt.legend(loc='upper left')
plt.savefig('ML_images/player_lr')
plt.show()
```
