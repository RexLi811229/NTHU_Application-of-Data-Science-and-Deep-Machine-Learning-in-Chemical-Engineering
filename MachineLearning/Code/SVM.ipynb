```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.svm import SVC
from utils import plot_decision_regions, get_player_data
```

```python
"""Kernel SVM toy example of 2d plot"""
np.random.seed(1)
# Set radius of the circle
r1 = 0.5
r1 = r1 * (1 + np.random.rand(100)) # Add noise
r2 = 2
r2 = r2 * (1 + np.random.rand(100)) # Add noise

# Generate random angles
alpha1 = 2 * np.pi * np.random.rand(100)
alpha2 = 2 * np.pi * np.random.rand(100)

# Calculating coordinates
x1 = r1 * np.cos(alpha1)
y1 = r1 * np.sin(alpha1)
x2 = r2 * np.cos(alpha2)
y2 = r2 * np.sin(alpha2)

plt.scatter(x1, y1,
            c='b',
            marker='x',
            label='1')
plt.scatter(x2, y2,
            c='r',
            marker='s',
            label='-1')
plt.legend(loc='upper left')
plt.savefig('ML_images/SVM_2d-example')
plt.show()
```

```python
"""Kernel SVM toy example of 3d plot"""
z1 = x1**2 + y1**2
z2 = x2**2 + y2**2

# Plot 3d scatter
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x1, y1, z1,
           c='b',
           marker='x',
           label='1')
ax.scatter(x2, y2, z2,
           c='r',
           marker='s',
           label='-1')

plt.legend(loc='upper left')
ax.view_init(10, 30) # Rotate the axes
plt.savefig('ML_images/SVM_3d-example')
plt.show()
```

```python
"""Gaussian kernel xor example of 2d plot"""
np.random.seed(1)
X_xor = np.random.randn(200, 2) # Give a normal distribution of array shaped (200, 2)
y_xor = np.logical_xor(X_xor[:, 0]>0, X_xor[:, 1]>0)
y_xor = np.where(y_xor, 1, -1)
plt.scatter(
    X_xor[y_xor==1, 0],
    X_xor[y_xor==1, 1],
    c='b',
    marker='x',
    label='1'
)
plt.scatter(
    X_xor[y_xor==-1, 0],
    X_xor[y_xor==-1, 1],
    c='r',
    marker='s',
    label='-1'
)
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.legend()
plt.savefig('ML_images/xor')
plt.show()
```

```python
"""Gaussian kernel xor example of 3d plot"""
svm_c = SVC(kernel='rbf', random_state=1, gamma=0.1, C=10.0)
svm_c.fit(X_xor, y_xor)
plot_decision_regions(X_xor, y_xor, classifier=svm_c)
plt.legend(loc='upper left')
plt.savefig('ML_images/xor_pred')
plt.show()
```

```python
"""Player position prediction by linear SVM"""
X_player_n, y_player = get_player_data()

svm_c = SVC(kernel='linear', random_state=1)
svm_c.fit(X_player_n, y_player)
plot_decision_regions(X_player_n, y_player, classifier=svm_c)
plt.legend(loc='upper left')
plt.savefig('ML_images/player_linear_SVM')
plt.show()
```

```python
"""Player position prediction by Gaussian kernel SVM"""
svm_c = SVC(kernel='rbf', random_state=1)
svm_c.fit(X_player_n, y_player)
plot_decision_regions(X_player_n, y_player, classifier=svm_c)
plt.legend(loc='upper left')
plt.savefig('ML_images/player_Gaussian_SVM')
plt.show()
```
