```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from utils import get_player_data, plot_decision_regions
```

```python
"""Player position prediction by random forest"""
X_player, y_player = get_player_data(normalized=False)

forest = RandomForestClassifier(criterion='entropy',
                                n_estimators=25,
                                random_state=1,
                                n_jobs=2,
                                min_samples_split=10,
                                max_depth=4,
                                max_leaf_nodes=20)
forest.fit(X_player, y_player)
plot_decision_regions(X_player, y_player, classifier=forest)

plt.legend(loc='upper left')
plt.savefig('ML_images/player_rf')
plt.show()

forest.feature_importances_
```

```python
"""Player position prediction by decision tree stumps"""
tree = DecisionTreeClassifier(criterion='entropy', random_state=1, max_depth=1)

tree.fit(X_player, y_player)
plot_decision_regions(X_player, y_player, classifier=tree)
plt.legend(loc='upper left')
plt.savefig('ML_images/player_dt_stumps')
plt.show()
```

```python
"""Player position prediction by adaboost"""
ada = AdaBoostClassifier(base_estimator=tree, n_estimators=25, random_state=1)
ada.fit(X_player, y_player)
plot_decision_regions(X_player, y_player, classifier=ada)
plt.legend(loc='upper left')
plt.savefig('ML_images/player_ada')
plt.show()
```
