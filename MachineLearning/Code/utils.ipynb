```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['xtick.labelsize'] = 14
plt.rcParams['ytick.labelsize'] = 14
```

```python
"""Create players `Pos` scatter plot with respect to height and weight"""
# Load players data
player_df = pd.read_csv('players.csv', index_col=False)

colors = ('b', 'r', 'g', 'k', 'c')
markers = ('.', 'o', 's', '*', 'x')
plt.figure(figsize=(12, 8))
skip_pos = ('F-C', 'G-F', 'F')
for pos, color, marker in zip(player_df['Pos'].unique(), colors, markers):
    if pos in skip_pos:
        continue
    x = player_df[player_df['Pos'] == pos]['Ht']
    y = player_df[player_df['Pos'] == pos]['Wt']
    plt.scatter(x=x, y=y, c=color, marker=marker, label=pos)

plt.title('Players Position of Varying Height and Weight', size=20)
plt.xlabel('Height (cm)', size=18)
plt.ylabel('Weight (kg)', size=18)
plt.legend()
plt.savefig('ML_images/players')
plt.show()
```
